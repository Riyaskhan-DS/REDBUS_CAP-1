```python
# importing libraries
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```


# Kerala Bus Detalis

```python
# Initialize WebDriver
driver = webdriver.Chrome()
driver.maximize_window()

# Read your CSV file with route links and names
df = pd.read_csv("kearala_links.csv")

# Lists to store the data
bus_data = {
    "route_link": [],
    "route_name": [],
    "Bus_Name": [],
    "Bus_Type": [],
    "Departing_Time": [],
    "Duration": [],
    "Reaching_Time": [],
    "Star_Rating": [],
    "Price": [],
    "Seat_Availability": []
}

for i, row in df.iterrows():
    link = row["route_link"]
    route = row["route_name"]

    driver.get(link)
    time.sleep(3)

    # Click to view bus details
    try:
        driver.find_element(By.XPATH, "//div[@class='button']").click()
        time.sleep(2)
    except Exception as e:
        print(f"Skipping link {link} due to an error: {e}")
        continue

    # Scroll to load all bus details
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(2)
        if driver.page_source == old_page_source:
            scrolling = False

    # Extract bus details
    bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_types = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_times = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_times = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    durations = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    ratings = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    prices = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    # Append data to lists
    for bus in bus_names:
        bus_data["route_link"].append(link)
        bus_data["route_name"].append(route)
        bus_data["Bus_Name"].append(bus.text)
    for bus_type in bus_types:
        bus_data["Bus_Type"].append(bus_type.text)
    for start in start_times:
        bus_data["Departing_Time"].append(start.text)
    for end in end_times:
        bus_data["Reaching_Time"].append(end.text)
    for duration in durations:
        bus_data["Duration"].append(duration.text)
    for rating in ratings:
        bus_data["Star_Rating"].append(rating.text)
    for price in prices:
        bus_data["Price"].append(price.text)
    for seat in seats:
        bus_data["Seat_Availability"].append(seat.text)


# Close the browser
driver.quit()
```

```python
KEARALA_BUS_DATA = pd.DataFrame(bus_data)
```

```python
# Initialize WebDriver
driver = webdriver.Chrome()
driver.maximize_window()

# Read your CSV file with route links and names
df = pd.read_csv("goa_links.csv")

# Lists to store the data
bus_data = {
    "route_link": [],
    "route_name": [],
    "Bus_Name": [],
    "Bus_Type": [],
    "Departing_Time": [],
    "Duration": [],
    "Reaching_Time": [],
    "Star_Rating": [],
    "Price": [],
    "Seat_Availability": []
}

for i, row in df.iterrows():
    link = row["route_link"]
    route = row["route_name"]

    driver.get(link)
    time.sleep(3)

   
    # Click on elements to reveal bus details
    elements = driver.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
        time.sleep(2) 

    
    # Scroll to load all bus details
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(2)
        if driver.page_source == old_page_source:
            scrolling = False

    # Extract bus details
    bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_types = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_times = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_times = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    durations = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    ratings = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    prices = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    # Append data to lists
    for bus in bus_names:
        bus_data["route_link"].append(link)
        bus_data["route_name"].append(route)
        bus_data["Bus_Name"].append(bus.text)
    for bus_type in bus_types:
        bus_data["Bus_Type"].append(bus_type.text)
    for start in start_times:
        bus_data["Departing_Time"].append(start.text)
    for end in end_times:
        bus_data["Reaching_Time"].append(end.text)
    for duration in durations:
        bus_data["Duration"].append(duration.text)
    for rating in ratings:
        bus_data["Star_Rating"].append(rating.text)
    for price in prices:
        bus_data["Price"].append(price.text)
    for seat in seats:
        bus_data["Seat_Availability"].append(seat.text)


# Close the browser
driver.quit()
```

```python
GOA_BUS_DATA = pd.DataFrame(bus_data)
```
