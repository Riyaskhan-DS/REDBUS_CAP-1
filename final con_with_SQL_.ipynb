```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
from decimal import Decimal
from datetime import datetime
import re
```

```python
# csv to dataframe
data1=pd.read_csv("scraped_bus_data_KEARALA.csv")
data2=pd.read_csv("scraped_bus_data_ANDHRA.csv")
data3=pd.read_csv("scraped_bus_data_GOA.csv")
data4=pd.read_csv("scraped_bus_data_HARYANA.csv")
data5=pd.read_csv("scraped_bus_data_RAJASTAN.csv")
data6=pd.read_csv("scraped_bus_data_SOUTHBENGAL.csv")
data7=pd.read_csv("scraped_bus_data_WESTBENGAL.csv")
data8=pd.read_csv("scraped_bus_data_ASSAM.csv")
data9=pd.read_csv("scraped_bus_data_TELUNGANA.csv")
data10=pd.read_csv("scraped_bus_data_UTTRAPRADESH.csv")

Final_data_REDBUS=pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10],ignore_index=True)
```

```python
Final_data_REDBUS
```

```python
# change dataframe to csv
Final_data_REDBUS.to_csv('FINAL_DATA_CAPSTONE_1.csv',index=False)
```

# Data cleaning

```python
# data about the data
Final_data_REDBUS.info()
```

```python
# price into decimal;

# Remove "INR" and convert prices to Decimal
Final_data_REDBUS["Price"] = Final_data_REDBUS["Price"].str.replace("INR", "").str.strip()

# Convert prices to Decimal
Final_data_REDBUS["Price"] = Final_data_REDBUS["Price"].apply(lambda x: Decimal(x) if x else Decimal(0))

# Fill NaN values with 0
Final_data_REDBUS["Price"] = Final_data_REDBUS["Price"].fillna(Decimal(0))
```

```python
# star_Rating to Float

# Remove "New" from Ratings
Final_data_REDBUS["Star_Rating"] = Final_data_REDBUS["Star_Rating"].str.replace("New", "")

# Strip any extra spaces
Final_data_REDBUS["Star_Rating"] = Final_data_REDBUS["Star_Rating"].str.strip()

# Take the first part (before any space) assuming it's the numeric rating
Final_data_REDBUS["Star_Rating"] = Final_data_REDBUS["Star_Rating"].str.split().str[0]

# Convert to float, coerce any invalid values (non-numeric) into NaN
Final_data_REDBUS["Star_Rating"] = Final_data_REDBUS["Star_Rating"].astype(float)

# Fill NaN values with 0
Final_data_REDBUS["Star_Rating"].fillna(0, inplace=True)
```

```python
Final_data_REDBUS.info()
```

```python
# replacing the nan value
Final_data_REDBUS = Final_data_REDBUS.replace({np.nan: None})
```

```python
# change dataframe to csv
Final_data_REDBUS.to_csv('CAPSTONE_1.CSV',index=False)
```

# Create a mysql database

```python
#sql connection
conn=mysql.connector.connect(host="localhost", user="root", password="root",database="redbus_data")
my_cursor = conn.cursor()
```

```python
# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS bus_routes(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  route_link VARCHAR(200),
                  route_name VARCHAR(200),
                  Bus_Name VARCHAR(150),
                  Bus_Type VARCHAR(150),
                  Departing_Time TIME,
                  Duration VARCHAR(150),
                  Reaching_Time TIME,
                  Star_Ratings Float,
                  Price decimal,
                  Seat_Availability VARCHAR(150)
                  )''')
print("Table Created successfully")
```

```python
insert_query = '''INSERT INTO bus_routes(
                  route_link,
                  route_name,
                  Bus_Name,
                  Bus_Type,
                  Departing_Time,
                  Duration,
                  Reaching_Time,
                  Star_Ratings,
                  Price,
                  Seat_Availability)
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''

data = REDBUS.values.tolist()

my_cursor.executemany(insert_query, data)

conn.commit()

print("Values inserted successfully")
```
